```python
import pandas as pd
from transformers import BertTokenizer
```

## 1. Import username data

```python
leftwing_users = pd.read_csv('./usernames/leftwing_users.csv')
leftwing_users
```

```python
rightwing_users = pd.read_csv('./usernames/patriots_win_users.csv')
rightwing_users
```

## 2. Tokenize Data

```python
# init tokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
```

```python
def preprocess_names(df):
    rez = []
    for index, row in df.iterrows():
        try:
            tokenized = tokenizer.tokenize(row[1])
            rez.append(tokenized)
        except:
            continue
    return rez
```

```python
rw_list = preprocess_names(rightwing_users)
rw_list[:5]
```

```python
lw_list = preprocess_names(leftwing_users)
lw_list
```
